```python
import json
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import pymongo

# Connecting to mongoDB
connection = pymongo.MongoClient('10.50.0.128', 27017)
db = connection.twitterdb

# Choosing the fields we want from the tweets
tweets = db.final_tweets3.find({"Bin": "Labour"}, {"text": 1, "polarity": 1, "subjectivity": 1, "filter": 1,"Bin": 1, "created_at": 1}) # pulling tweets from mongo

# Create a list for each tweet in tweets
l = [t for t in tweets]

# Create a data frame from the list of tweets
df = DataFrame(l)
```

```python
# Add column that returns true if desired word is in tweet
```

```python
df['education'] = df['text'.lower()].str.contains('education')
```

```python
df['tax'] = df['text'.lower()].str.contains('tax')
```

```python
df['health'] = df['text'.lower()].str.contains('health')
```

```python
# change the date to an appropriate datetime format
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df['created_at'] = df['created_at'].values.astype('datetime64[h]')
```
